```python
import pandas as pd
import numpy as np

# Creating project_data dictionary
project_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005", "A002", "A005", "A003", "A001", "A003", "A001", "A004", "A005", "A005"],
    "Project": ["Project 1", "Project 2", "Project 3", "Project 4", "Project 5", "Project 6", "Project 7", 
                "Project 8", "Project 9", "Project 10", "Project 11", "Project 12", "Project 13", "Project 14"],
    "Cost":[1002000, 2000000, 4500000, 5500000, np.nan,680000, 400000, 350000, np.nan,300000, 2000000, 1000000, 
             3000000, 200000],
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", 
               "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished"]
}

# Creating the DataFrame
project_df = pd.DataFrame(project_data)

#Displaying the Dataframe
(project_df)
```

```python
import pandas as pd
#Creating employee Dataframe

Employee_dataframe={ 
    "ID":["A001", "A002", "A003", "A004", "A005"],
    "Full Name":["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johnny"],
    "Gender":["M","F","M","F","F"],
    "City":["Paris","London","Berlin","Newyork","Madrid"],
    "Age":[25,27,29,31,30]
}

employee_df=pd.DataFrame(Employee_dataframe)

# Displaying the DataFrame
(employee_df)
```

```python
import pandas as pd

#Creating Seniority_level_dataframe

Seniority_level_dataframe={
          "ID":["A001", "A002", "A003", "A004", "A005"],
          "Designation Level":[2,2,3,2,3]
}

# Creating the DataFrame
seniority_df=pd.DataFrame(Seniority_level_dataframe)

 #Displaying DataFrame
(seniority_df)
```

```python
#Saving the dataframe in CSV Files

project_df.to_csv("Project_data.csv",index=False)
employee_df.to_csv("Employee_data.csv",index=False)
seniority_df.to_csv("Seniority_level_dataframe.csv",index=False)

print("DataFrames have been successfully saved as CSV files!")
```

```python

# Initialize variables for running sum and valid count
running_sum = 0
valid_count = 0

# Iterate through each row and fill missing 'Cost' values
for i in range(len(project_df)):
    if pd.isnull(project_df.loc[i, "Cost"]):  # Check if 'Cost' is NaN
        if valid_count > 0:
            project_df.loc[i, "Cost"] = running_sum / valid_count  # Replace with running average
    else:
        running_sum += project_df.loc[i, "Cost"]  # Add to running sum
        valid_count += 1  # Increment valid count

# Disable scientific notation for better readability
pd.set_option('display.float_format', '{:.0f}'.format)


# Display the updated DataFrame
(project_df)
```

```python
# Extracting names and splitting them
names = list(employee_df['Full Name'])

# Splitting the names into first and second parts
first_names = list(map(lambda x: x.split()[0], names))
last_names = list(map(lambda x: x.split()[1], names))

# Inserting new columns "First Name" and "Last Name"
employee_df.insert(1, "First Name", first_names)
employee_df.insert(2, "Last Name", last_names)

# Dropping the original "Name" column
employee_df.drop('Full Name', inplace=True, axis=1)

# Display the updated DataFrame
(employee_df)
```

```python
#Merging the Seniority dataframe with the Employee dataframe 

pre_final_df = pd.merge(employee_df, seniority_df, on="ID")
pre_final_df
```

```python
#Creating a final dataframe with the needed data of the employees

final = pd.merge( pre_final_df, project_df, on='ID')
final
```

```python
final['Bonus-5%'] = final['Cost'].where(final['Status'] == "Finished", 0) * 0.05

# Display the updated DataFrame
(final)
```

```python
# Function to decrease designation level by 1 if project status is "Fail"
def decrease_designation(row):
    if row['Status'] == 'Failed':
        # Decrease designation level by 1, but ensure it doesn't go below 1
        return max(row['Designation Level'] - 1, 1)
    return row['Designation Level']

# Apply the function to the DataFrame to update the Designation Level
final['Updated Designation Level'] = final.apply(decrease_designation, axis=1)

# Display the updated DataFrame
(final[['ID', 'Project', 'Status', 'Designation Level', 'Updated Designation Level']])
```

```python
# Adding "Mr." or "Mrs." to the first name based on gender
employee_df["First Name"] = employee_df.apply(
    lambda row: ("Mr. " if row["Gender"] == "M" else "Mrs. ") + row["First Name"],
    axis=1
)

# Dropping the 'Gender' column
employee_df = employee_df.drop(columns=["Gender"])

# Display the updated DataFrame
employee_df
```

```python
# Function to promote designation level by 1 if age > 29
def promote_designation(row):
    if row['Age'] > 29:
        return row['Designation Level'] + 1  # Increase by 1
    return row['Designation Level']  # Keep the same

# Apply the function to create a new 'Promoted Designation Level' column
final['Promoted Designation Level'] = final.apply(promote_designation, axis=1)

final[['ID', 'Age', 'Designation Level', 'Promoted Designation Level']]

```

```python
# Group by 'ID' and 'First Name', then calculate the total cost
TotalProjCost = final.groupby(['ID', 'First Name'], as_index=False)['Cost'].sum()

# Rename the columns
TotalProjCost.rename(columns={'Cost': 'Total Cost'}, inplace=True)

#Print the new DataFrame
(TotalProjCost)
```

```python
# Filter rows where 'City' contains the letter 'o'
employees_with_o_in_city = final[final['City'].str.contains('o', case=False)]

# Display the filtered DataFrame
(employees_with_o_in_city)
```
